**14 -** Create a stacked bar plot using plotnine with the following info:
- x = publication year
- y = count of instrument models

The bars should be stacked, where each "subbar" represents an instrument

```python
plot_df = lib_df.groupby(by="publication_year")["instrument_model"].value_counts().to_frame().reset_index()

(ggplot(plot_df, aes(plot_df["publication_year"].astype(str), "count", fill="instrument_model")) +
 geom_col(color="black") +
 theme_classic() +
 theme(figure_size=(14, 6)) +
 scale_y_continuous(expand=(0, 0)))
```

**15 -** How many single and double stranded samples are there, grouped by publication year? 

Hint: a bar chart can show this very nicely

```python
plot_df = lib_df.groupby(by="publication_year")["strand_type"].value_counts().reset_index()

(ggplot(plot_df, aes("publication_year", "count", fill="strand_type")) +
 geom_col(color="black") +
 scale_y_continuous(expand=(0,0)))
```
